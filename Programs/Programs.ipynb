```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Set plot style for consistency
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Default figure size

# --- 1. Dataset Simulation based on the Head and Correlation Matrix ---

# Define the number of rows for the simulated dataset
N = 1000
np.random.seed(42) # For reproducibility

# 1. Simulate Highly Correlated Features (SpendingScore & Target)
# The correlation heatmap shows: SpendingScore vs Target: 0.63
# We'll use a multivariate normal distribution to enforce this correlation.
mean = [0, 0]
cov_matrix = [[1, 0.63], [0.63, 1]]
correlated_data = np.random.multivariate_normal(mean, cov_matrix, N)

# Scale and transform to desired ranges/distributions
# SpendingScore (Range 0-100)
SpendingScore = np.clip(correlated_data[:, 0] * 20 + 50, 0, 100)

# Target (Range 100-250, Mean around 175)
Target = np.clip(correlated_data[:, 1] * 25 + 175, 100, 250)

# 2. Simulate other numerical features based on typical ranges
# Age (Range 18-70)
Age = np.random.randint(18, 71, N)

# Income (Range 20000-100000, slightly right-skewed)
Income = np.clip(np.random.lognormal(mean=10.9, sigma=0.4, size=N), 20000, 100000)

# NumPurchases (Range 1-10, mostly low)
NumPurchases = np.random.poisson(lam=4, size=N)
NumPurchases = np.clip(NumPurchases, 1, 10)

# CreditScore (Range 600-800, normally distributed)
CreditScore = np.clip(np.random.normal(loc=700, scale=40, size=N), 600, 800)

# SatisfactionScore (Range 0-10)
SatisfactionScore = np.random.uniform(3, 10, N)

# 3. Simulate Binary Features
# IsSubscribed (Binary: 0 or 1, correlation with Target: 0.21)
IsSubscribed_prob = 0.5 + 0.001 * (Target - Target.mean()) # Introduce slight correlation
IsSubscribed = (np.random.rand(N) < IsSubscribed_prob).astype(int)

# HasCreditCard (Binary: 0 or 1)
HasCreditCard = np.random.randint(0, 2, N)

# YearsWithCompany (Range 1-40, mostly older employees)
YearsWithCompany = np.random.gamma(shape=5, scale=6, size=N)
YearsWithCompany = np.clip(YearsWithCompany, 1, 40)

# 4. Simulate Categorical Features
# Gender
Gender = np.random.choice(['Male', 'Female'], N, p=[0.5, 0.5])

# City
City = np.random.choice(['CityA', 'CityB', 'CityC'], N, p=[1/3, 1/3, 1/3])

# MaritalStatus
MaritalStatus = np.random.choice(['Married', 'Single', 'Divorced'], N, p=[0.5, 0.3, 0.2])

# 5. Create the DataFrame
df = pd.DataFrame({
    'Age': Age.astype(int),
    'Income': Income,
    'SpendingScore': SpendingScore,
    'NumPurchases': NumPurchases.astype(int),
    'CreditScore': CreditScore,
    'Gender': Gender,
    'City': City,
    'IsSubscribed': IsSubscribed,
    'MaritalStatus': MaritalStatus,
    'YearsWithCompany': YearsWithCompany.astype(int),
    'HasCreditCard': HasCreditCard,
    'SatisfactionScore': SatisfactionScore,
    'Target': Target
})

# Adjust data types
df['Income'] = df['Income'].round(2)
df['SpendingScore'] = df['SpendingScore'].round(2)
df['CreditScore'] = df['CreditScore'].round(2)
df['SatisfactionScore'] = df['SatisfactionScore'].round(2)
df['Target'] = df['Target'].round(2)

# --- 2. Print DataFrame Head (First Output Image) ---
print("--- Data Head (Simulation) ---")
print(df.head().to_markdown(index=True))
print("\n" + "="*50 + "\n")


# --- 3. Generate Visualizations ---

# Identify numerical features for pairplot and correlation
numerical_features = ['Age', 'Income', 'SpendingScore', 'NumPurchases', 'CreditScore',
                      'IsSubscribed', 'YearsWithCompany', 'HasCreditCard',
                      'SatisfactionScore', 'Target']
# Select features that appear in the PairPlot and Correlation Heatmap
heatmap_cols = ['Age', 'Income', 'SpendingScore', 'NumPurchases', 'CreditScore',
                'IsSubscribed', 'YearsWithCompany', 'HasCreditCard',
                'SatisfactionScore', 'Target']
pairplot_cols = ['Age', 'Income', 'SpendingScore', 'NumPurchases', 'CreditScore',
                 'SatisfactionScore', 'Target'] # Features explicitly shown in PairPlot image

## 3.1 Scatterplot Matrix (PairPlot) of Numerical Features
plt.figure(figsize=(10, 10))
sns.pairplot(df[pairplot_cols], diag_kind='kde', corner=True)
plt.suptitle('Scatterplot Matrix of Numerical Features', y=1.02)
plt.show()
print("="*50 + "\n")


## 3.2 PCA Plot (Income vs SpendingScore and PC1 vs PC2)
plt.figure(figsize=(16, 6))
# A. Before PCA: Income vs SpendingScore
plt.subplot(1, 2, 1)
sns.scatterplot(x='Income', y='SpendingScore', hue='City', data=df, alpha=0.6, palette='Set2')
plt.title('Before PCA: Income vs SpendingScore')
plt.grid(True)
plt.xlabel('Income')
plt.ylabel('Spending Score')

# B. After PCA: PC1 vs PC2
# 1. Select and scale the numerical features
features = ['Income', 'SpendingScore']
X = df[features].values
X_scaled = StandardScaler().fit_transform(X)

# 2. Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2'])
pca_df['City'] = df['City'].values

# 3. Plot PCA results
plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='City', data=pca_df, alpha=0.6, palette='Set2')
plt.title('After PCA: PC1 vs PC2')
plt.grid(True)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')

plt.tight_layout()
plt.show()
print("="*50 + "\n")


## 3.3 Correlation Heatmap of Numerical Features
plt.figure(figsize=(12, 8))
# Calculate the correlation matrix
corr_matrix = df[heatmap_cols].corr().round(2)
# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
print("="*50 + "\n")


## 3.4 Parallel Coordinates Plot by Gender
plt.figure(figsize=(14, 7))
# Prepare data: Normalize features for the parallel plot
data_to_plot = df[['Age', 'Income', 'SpendingScore', 'NumPurchases', 'CreditScore', 'SatisfactionScore', 'Gender']].copy()
# Normalize selected columns (MinMaxScaler equivalent for plotting)
for col in data_to_plot.columns.drop('Gender'):
    min_val = data_to_plot[col].min()
    max_val = data_to_plot[col].max()
    data_to_plot[col] = (data_to_plot[col] - min_val) / (max_val - min_val)

# Create the plot
from pandas.plotting import parallel_coordinates
# Use the same colors as the image (approximately: cyan for Male, magenta for Female)
parallel_coordinates(data_to_plot, 'Gender', color=['cyan', 'magenta'])
plt.title('Parallel Coordinates Plot by Gender')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
print("="*50 + "\n")


## 3.5 Average Target by Age
plt.figure(figsize=(10, 5))
# Calculate the mean target for each age
target_by_age = df.groupby('Age')['Target'].mean().reset_index()
sns.lineplot(x='Age', y='Target', data=target_by_age, marker='o', errorbar=None)
plt.title('Average Target by Age')
plt.grid(True)
plt.show()
print("="*50 + "\n")


## 3.6 Subscription Status across Cities (Stacked Bar Plot)
plt.figure(figsize=(8, 6))
# Create a cross-tabulation of City and IsSubscribed
city_subscription = pd.crosstab(df['City'], df['IsSubscribed'])
# Plot as a stacked bar chart
city_subscription.plot(kind='bar', stacked=True, color=['#4B0082', '#FFFF00'], ax=plt.gca(), width=0.5)
plt.title('Subscription Status across Cities')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='IsSubscribed', labels=['0', '1']) # 0: Not Subscribed, 1: Subscribed
plt.grid(axis='y') # Only horizontal grid lines
plt.show()
print("="*50 + "\n")


## 3.7 Concentric/Sunburst Chart (Simulation of Regions and Categories)
# Note: The data for this complex chart (Regions, Categories like Clothing, Snacks, Men/Women)
# is not present in the dataframe head. This is a purely visual simulation.

# Create dummy data for a sunburst chart structure
data_sunburst = {
    'Region': ['North', 'East', 'South', 'West'] * 5,
    'Category': ['Grocery', 'Clothing', 'Electronics', 'Snacks', 'Beverages'] * 4,
    'SubCategory': ['Kids', 'Women', 'Men', 'Laptops', 'Phones'] * 4,
    'Sales': np.random.randint(10, 100, 20)
}
sunburst_df = pd.DataFrame(data_sunburst)

# Since Matplotlib/Seaborn doesn't natively support sunburst charts easily,
# we'll just print a placeholder message for the user, acknowledging the complexity.
print("Concentric Chart (Sunburst Chart):")
print("> Generating this specific concentric chart requires advanced plotting libraries (e.g., Plotly, or complex Matplotlib/Dendrogram code) and the original categorical data (Regions, Subcategories) which is not available in the simulated DataFrame.")
print("> The image represents sales breakdown by **Region** (inner ring), **Category** (middle ring), and **Subcategory** (outer ring).")

# You would typically use Plotly for this type of visualization:
# import plotly.express as px
# fig = px.sunburst(sunburst_df, path=['Region', 'Category', 'SubCategory'], values='Sales')
# fig.show()
```

```python
# Force-based Layout Visualization using NetworkX
import networkx as nx
import matplotlib.pyplot as plt

#---- Step 1: Create sample dataset (edges)
edges = [
    ("Alice", "Bob"),
    ("Alice", "Cathy"),
    ("Alice", "David"),
    ("Bob", "Eve"),
    ("Cathy", "Frank"),
    ("David", "Grace"),
    ("Eve", "Helen"),
    ("Frank", "Grace"),
    ("Helen", "Ivy"),
    ("Grace", "Jack"),
    ("Ivy", "Jack")
]

#---- Step 2: Create a Graph ----
G = nx.Graph()
G.add_edges_from(edges)

#---- Step 3: Compute Force-based Layout ---
# spring_layout uses the Fruchterman-Reingold algorithm
pos = nx.spring_layout(G, seed=42) # Using a fixed seed for reproducible layout

#---- Step 4: Visualize the Graph ----
plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True,
        node_color='lightblue',
        node_size=1200,
        font_size=10,
        font_weight='bold',
        edge_color='gray')

plt.title("Force-based Layout Visualization", fontsize=14)
plt.show()
```

```python
# Enable inline plots in Colab
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import networkx as nx
from itertools import combinations

# --- 1. Source text recreated ---
text_words = {
    'systems': 50, 'data': 45, 'artificial': 40, 'ai': 35,
    'intelligence': 30, 'transforming': 25, 'analyzing': 25, 'live': 20,
    'large': 20, 'capable': 18, 'work': 18, 'continuous': 15, 'improving': 15,
    'faster': 15, 'patterns': 12, 'decisions': 12, 'making': 12,
    'healthcare': 10, 'finance': 10, 'efficiency': 10, 'accuracy': 10,
    'build': 8, 'whole': 8, 'research': 8, 'applications': 8, 'concerns': 8,
    'challenges': 8, 'privacy': 8, 'ethical': 8, 'bias': 8, 'transparency': 8,
    'remain': 8, 'trustworthy': 6, 'benefit': 6, 'innovation': 6, 'society': 6,
    'helping': 6, 'way': 6, 'humans': 6, 'amounts': 6, 'critical': 6,
    'recognizing': 6
}

# ✅ Correct text expansion
full_text = ' '.join(word for word, freq in text_words.items() for _ in range(freq))

# --- 2. Word Cloud ---
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    if word in ['ai', 'data', 'systems']:
        return "rgb(255, 193, 7)"  # Gold
    elif font_size > 40:
        return "rgb(94, 19, 142)"  # Purple
    elif font_size > 25:
        return "rgb(205, 92, 92)"  # Indian Red
    else:
        return "rgb(0, 0, 128)"  # Navy

wordcloud = WordCloud(
    width=1000, height=500, background_color='white',
    max_words=100, random_state=42
).generate(full_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud.recolor(color_func=custom_color_func, random_state=42), interpolation='bilinear')
plt.title("Word Cloud - Artificial Intelligence Text", fontsize=16)
plt.axis('off')
plt.show()

# --- 3. Network Graph ---
co_occurrence_groups = [
    ['improving', 'applications', 'finance', 'efficiency', 'accuracy', 'healthcare'],
    ['concerns', 'challenges', 'bias', 'ethical', 'privacy', 'transparency', 'remain'],
    ['society', 'continuous', 'helping', 'innovation', 'trustworthy', 'benefit', 'whole', 'research', 'build'],
    ['artificial', 'way', 'live', 'work', 'intelligence', 'transforming'],
    ['humans', 'capable', 'decisions', 'faster', 'making', 'recognizing', 'analyzing', 'amounts'],
    ['ai'], ['data'], ['systems']
]

G = nx.Graph()
for group in co_occurrence_groups:
    G.add_edges_from(combinations(group, 2))
G.add_edges_from([('ai', 'data'), ('ai', 'systems')])

pos = nx.spring_layout(G, k=0.8, iterations=50, seed=42)

plt.figure(figsize=(10, 8))
nx.draw(
    G, pos, with_labels=True, node_size=1200, node_color='lightblue',
    font_size=8, font_color='black', edge_color='gray', width=1.5, alpha=0.8
)
plt.title("Text Network Visualization - Word Co-occurrence Graph", fontsize=14)
plt.show()
```

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns, plotly.express as px

sns.set_style("darkgrid"); plt.rcParams['figure.figsize'] = (14, 6)
np.random.seed(42)

# --- 1. Data Simulation ---
states = ['Alabama','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','Florida','Georgia','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming']

region_map = {**dict.fromkeys(['California','Washington','Oregon','Nevada','Arizona','Utah','Idaho','Montana','Wyoming','New Mexico','Colorado'],'West'),
              **dict.fromkeys(['New York','Pennsylvania','New Jersey','Massachusetts','Maryland','Connecticut','Rhode Island','Delaware','Maine','New Hampshire','Vermont'],'East'),
              **dict.fromkeys(['Texas','Illinois','Michigan','Ohio','Indiana','Wisconsin','Iowa','Kansas','Nebraska','Minnesota','North Dakota','South Dakota','Missouri'],'Central')}
region_map = {s: region_map.get(s,'South') for s in states}

top_targets = {'California':457687.63,'New York':310876.27,'Texas':170188.05,'Washington':138641.27,'Pennsylvania':116511.91}
weights = np.array([top_targets.get(s,10000)/1e5 for s in states]); weights /= weights.sum()

N, SALES = 10000, 230
df = pd.DataFrame({'State': np.random.choice(states, N, p=weights)})
df['Region'] = df['State'].map(region_map)
df['Sales'] = np.random.lognormal(mean=np.log(SALES), sigma=0.8, size=N)
state_sales = df.groupby('State', as_index=False)['Sales'].sum().rename(columns={'Sales':'Total_Sales'})
df = df.merge(state_sales, on='State')

# --- 2. Summary ---
total_sales = state_sales['Total_Sales'].sum()
top5 = state_sales.nlargest(5, 'Total_Sales').reset_index().rename(columns={'index':'Index'})
top5['Index'] += 1
region_sum = df.groupby('Region')['Sales'].agg(['sum','mean']).reset_index()

print(f"--- SUMMARY ---\nTotal States: {df['State'].nunique()}, Total Sales: ${total_sales:.2f}\n\nTop 5 States:")
print(top5.to_string(index=False, float_format='{:.4f}'.format))
print("\nSales by Region:")
print(region_sum.to_string(index=False, float_format='{:.6f}'.format))
print("\n" + "="*50 + "\n")

# --- 3. Visualizations ---

## 3.1 Choropleth Map
fig = px.choropleth(state_sales, locations='State', locationmode='USA-states', color='Total_Sales',
                    scope='usa', color_continuous_scale='Blues', title='Superstore Sales by State')
fig.update_layout(coloraxis_colorbar=dict(title='Total_Sales', tickvals=[1e5,2e5,3e5,4e5],
                                           ticktext=['100k','200k','300k','400k']))
fig.write_json("Superstore_Sales_Choropleth.json")
print("Choropleth Map generated and saved as Superstore_Sales_Choropleth.json\n" + "="*50 + "\n")

## 3.2 Top 10 & Region Sales
fig, ax = plt.subplots(1,2,figsize=(16,6))
sns.barplot(x='Total_Sales', y='State', data=state_sales.nlargest(10,'Total_Sales'),
            color='steelblue', ax=ax[0])
ax[0].set(title='Top 10 States by Sales', xlabel='', ylabel='')

sns.barplot(x='Region', y='sum', data=region_sum, palette='Set2', ax=ax[1])
ax[1].set(title='Sales by Region', xlabel='Region', ylabel='sum', ylim=(0,750000))
plt.show(); print("Top 10 States and Sales by Region bar plots generated.\n" + "="*50 + "\n")

## 3.3 Heatmap
rows = ['Alabama','Arkansas','Colorado','Delaware','Florida','Idaho','Indiana','Kansas','Louisiana','Maryland','Michigan','Mississippi','Montana','Nevada','New Jersey','New York','North Dakota','Oklahoma','Pennsylvania','South Carolina','Tennessee','Utah','Virginia','West Virginia','Wyoming']
hm = df.groupby(['State','Region'])['Sales'].sum().unstack(fill_value=0)
sns.heatmap(hm.loc[hm.index.intersection(rows)], cmap='YlOrRd', linewidths=.5, linecolor='white',
            cbar_kws={'label':'Sales','ticks':[0,1e5,2e5,3e5,4e5]})
plt.title("Sales Heatmap: State vs Region"); plt.ylabel("State"); plt.xlabel("Region"); plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Setup ---
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (16, 6)
np.random.seed(42)

# --- 2. Simulate Data ---
dates = pd.date_range('2014-01-01', '2016-01-01', freq='D')
N = len(dates)
month, year = dates.month, dates.year

sales = (
    (10 + np.random.normal(0, 1.5, N)) +
    (np.sin(2*np.pi*month/12)*3 + np.cos(2*np.pi*month/6)*1.5) +
    np.where(year == 2015, 1.5, 0.5)
) * 10
sales = np.clip(sales, 10, 300)

items = [f"item{i:02d}" for i in range(50)] + [
    'whole milk','other vegetables','rolls/buns','soda','yogurt',
    'bottled water','pastry','citrus fruit','sausage','shopping bags',
    'semi-finished bread','bottled beer','root vegetables','tropical fruit',
    'canned fish','mustard','flower (seeds)','rubbish bag','salt'
]

df = pd.DataFrame({'Date': dates, 'Sales': sales})
df = df.loc[df.index.repeat(np.random.randint(1, 4, N))]
df['itemDescription'] = np.random.choice(items, len(df))
df['month'], df['year'] = df['Date'].dt.month, df['Date'].dt.year

# --- 3. Aggregates ---
df_daily = df.groupby('Date', as_index=False)['Sales'].sum()
df_daily['Monthly_Avg'] = df_daily['Sales'].rolling(30, center=True).mean()

df_monthly = df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum().reset_index()
df_monthly['Sales'] *= 50
df_monthly['Date'] = df_monthly['Date'].astype(str)

df_yearly = df.groupby(df['Date'].dt.year)['Sales'].sum().reset_index()
df_yearly.columns = ['Year', 'Yearly Sales']
df_yearly['Yearly Sales'] *= 12

heatmap_data = (
    df.groupby([df['year'], df['month']])['Sales'].sum().unstack() * 50
)

# --- 4. Plots ---
## 4.1 Daily vs Monthly Trend
plt.plot(df_daily['Date'], df_daily['Sales'], color='lightcoral', lw=5, alpha=0.5, label='Daily Sales')
plt.plot(df_daily['Date'], df_daily['Monthly_Avg']*20, 'r--', lw=2, label='Monthly Avg')
plt.title("Sales Trend Over Time"); plt.legend(); plt.grid(); plt.show()

## 4.2 Monthly + Yearly Sales
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# A. Monthly
time_index = np.linspace(2014.0, 2016.0, len(df_monthly))
ax[0].plot(time_index, df_monthly['Sales'], color='green', lw=2)
ax[0].set(title="Monthly Sales", ylim=(1400,1950))
xt = np.arange(2014.0, 2016.25, 0.25)
ax[0].set_xticks(xt)
ax[0].set_xticklabels([f"{int(x)}-{int((x*12)%12)+1:02d}" if x%1 else f"{int(x)}" for x in xt])
ax[0].grid()

# B. Yearly
ax[1].bar(df_yearly['Year'], df_yearly['Yearly Sales'], color='orange', width=0.5)
ax[1].set(title="Yearly Sales", ylim=(0,21000))
ax[1].grid()
plt.show()

## 4.3 Monthly Distribution + Heatmap
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

sns.boxplot(
    x='month', y='itemDescription', data=df[df['itemDescription'].isin(items)],
    color='lightgray',
    flierprops={'marker':'o','markersize':5,'markerfacecolor':'white','markeredgecolor':'black'},
    ax=ax[0]
)
ax[0].set(title="Monthly Distribution", xlabel="month", ylabel="itemDescription")

sns.heatmap(
    heatmap_data, cmap='YlGnBu', vmin=1400, vmax=1950,
    linewidths=0.5, linecolor='white',
    cbar_kws={'label':'Sales','ticks':[1500,1600,1700,1800,1900]}, ax=ax[1]
)
ax[1].set(title="Sales Heatmap (Year vs Month)", xlabel="month", ylabel="year", yticks=[0,1])
plt.show()
```
